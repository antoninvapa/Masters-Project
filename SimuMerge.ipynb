```python
import numpy as np
#temps de merge
time_step=1000
#temps de simulation
T_tot = 5000
#nombre d'étoiles
N_initial = 217
#nombre d'expérience
N_exp=1000


t = np.arange(0, T_tot, time_step)

#liste d'accueil de mes resultats de simu
b, s, n= np.empty(N_exp), np.empty(N_exp), np.empty(N_exp)

for k in range(1000):
    #loop sur mes experience en initiant les conditions
    B=0
    S=0
    N=N_initial
    for i in range(len(t)):
        #loop sur chaque etape de merge
        B=0
        S=0

        for j in range(N):
            #loop sur toutes les étoiles
            c=np.random.choice((0, 1))

            if  c==1:
                #l'étoile a un companion
                B+=1
            
            else:
                #l'étoile est toute seule la pauvre
                S+=1
            
            if 2*B+S==N: 
                #condition pour pas compter deux fois les étoiles à companion
                break
        #après time_step les binaries system mergent et donc le nombre final d'étoile réduit
        N=B+S
    
    #à la fin de la loop de temps on ajoute les résultats sur les listes d'accueil et on recommence N_exp fois
    b[k]=B
    s[k]=S
    n[k]=S+2*B
```

```python
print(np.mean(b), np.mean(s), np.mean(n), '\n', b, s, n)
```
